```python
import numpy as np 
# -------------- LISBON GENERATOR ------------------------
# or_tw = [ i*15 for i in range(8)  ]
# dur_ = [30,45, 60]
all_requests = []
for or_point in range(dem_mat.shape[0]):
    temp_or_vec = np.where(dem_mat.iloc[or_point,:] == 1)[0].tolist()
    # print(samp_or)
    if len(temp_or_vec)>0:
        for onedest in temp_or_vec:

            num_of_pass = dem_mat.iloc[or_point,onedest] 

            # print('destination',dest_samp)

            or_time  = 0

            duration  = 20

            dest_tw  = or_time + duration 
            if or_point == onedest or or_point == 0:
                continue
            res = {'or_point': or_point, 'dest_point': onedest, 'start_time': or_time,
            'dest_time' : dest_tw, 'num_people': num_of_pass}

            all_requests.append(res)
print(all_requests)
```

```python
import numpy as np 
or_tw = [ i*15 for i in range(8)]
dur_ = [30,45, 60]
all_requests = []
for _ in range(10):
    samp_or = np.random.choice(18,p =  np.array(dep)/np.sum(dep))
    # print(samp_or)
    num_of_pass = np.random.poisson(dep[samp_or], 1)
    
    if num_of_pass == 0:
        num_of_pass = [1]

    # print('number',num_of_pass)

    temp_arr_  = arr.copy()
    temp_arr_[samp_or] = 0 

    dest_samp = np.random.choice(18,p = np.array(temp_arr_)/np.sum(temp_arr_))

    # print('destination',dest_samp)

    or_time  = np.random.choice(or_tw)

    duration  = np.random.choice(dur_)

    dest_tw  = or_time + duration 

    res = {'or_point': samp_or, 'dest_point': dest_samp, 'start_time': or_time,
    'dest_time' : dest_tw, 'num_people': num_of_pass[0]}

    all_requests.append(res)
print(all_requests)
```

```python
import uuid 
import  datetime 

aimsun_reqs_dict = []
ids_re = []
count = 0
for one_req in all_requests: 
    # temp_dict = {}
    or_time = one_req['start_time']
    or_hour = or_time//60 + 8 
    or_min = or_time%60 

    dest_hour = or_time//60 + 8 
    dest_min = or_time%60 

    dest_time = one_req['dest_time']
    dtime = int(datetime.datetime(2019, 10, 26, or_hour, or_min).timestamp()*1000 )
    for _ in range(one_req['num_people']):
        temp_dict = {}
        temp_dict['time'] = dtime
        temp_dict['destination_time'] = float((dest_time - or_time)*60*1000)
        temp_dict['id'] = uuid.uuid4().__str__()
        temp_dict['name'] = 'request_{0}'.format(count)
        temp_dict['person_type'] = 'Type2CurrentModeNonCar'
        temp_dict['req_purpose'] = 'Other'
        temp_dict['origin'] = {'object':'Station_{0}'.format(one_req['or_point']) , 'stop_time': 300}
        temp_dict['destination'] = {'object':'Station_{0}'.format(one_req['dest_point']) , 'stop_time':300 }
        temp_dict['people'] = [{"cargo_space": 1}]
        temp_dict['availability'] = 1
        temp_dict['maxWalkingTime'] = 600
        count +=1
        # aimsun_reqs_dict.append(json.dumps(temp_dict))
        aimsun_reqs_dict.append(temp_dict)
        ids_re.append(temp_dict['id'])

aimsun_reqs_dict
# ids_re
```

```python
from sklearn.metrics.pairwise import haversine_distances
from math import radians

def getHaveDistance(coords): 

    coords_or = [[d1,d2] for d1,d2 in zip(coords.lat.tolist(), coords.lon.tolist())]

    # or_list = zip(or_lat,or_lon)
    # dest_list = [des_lat, des_lon]

    or_in_radians = [[radians(_) for _ in a1] for a1 in coords_or ] 
    # dest_in_radians = [radians(_) for _ in dest_list]

    result = haversine_distances(or_in_radians, or_in_radians)
    result = result * 6371000/1000 

    return result, result*2

dist, time = getHaveDistance(fdf)
time.tolist()

len(time)
```

```python
import json 
class MyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(MyEncoder, self).default(obj)

with open('AIMSUN_FILES_TEST\darp_input_lisbon_s2.json', 'w') as of: 
    json.dump(indata, of, indent= 2, cls=MyEncoder)
```
